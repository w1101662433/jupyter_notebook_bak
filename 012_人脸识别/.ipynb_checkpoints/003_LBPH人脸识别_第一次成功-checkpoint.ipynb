```python
from mtcnn import MTCNN
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os,sys,gc
from torch.utils.data import Dataset 
import os
from PIL import Image
from torchvision.transforms import ToTensor
import torch
import torchvision
from torch import nn 
from torch.nn.functional import normalize
from matplotlib.patches import Rectangle
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定中文字体，如SimHei
plt.rcParams['axes.unicode_minus'] = False  # 解决负号'-'显示为方块的问题
```

```python
Base_path = r"D:\BaiduSyncdisk\jupyter\012_人脸识别\dataset\生成的人脸"
all_labels = sorted([label for label in os.listdir(Base_path)])
detector = MTCNN()
```

```python
features = []
labels = []

def create_train():
    # loop every folder
    for person in all_labels:
        if person not in ['刘亦菲','冯提莫','杨颖',]:
            continue
        path = os.path.join(Base_path, person)
        label = all_labels.index(person)
        # loop every image in the folder
 
        for img in os.listdir(path):
            img_path = os.path.join(path, img)
            # read the image from the path
             
            img_array  = cv2.imdecode(np.fromfile(img_path, dtype=np.uint8), -1)

            try:
                gary = cv2.cvtColor(img_array, cv2.COLOR_BGR2GRAY)
            except:
                continue
            
            features.append(gary)
            labels.append(label)

create_train()
print('create_data success ------------------')
 
```

```python
# 训练
features = np.array(features,dtype='object')
labels = np.array(labels)
 
face_recognizer = cv2.face.LBPHFaceRecognizer_create()
 
# Train the Recognizer on the features list and the labels list
face_recognizer.train(features,labels)
# Pass two parameters: 1.scr--->the target image 2.label---->the corresponding label of the image
face_recognizer.save('face_trained.yml')
np.save('features.npy',features)
np.save('labels.npy',labels)
```

```python
# 加载识别器
face_recognizer = cv2.face.LBPHFaceRecognizer_create()
face_recognizer.read('face_trained.yml')
features = np.load('features.npy', allow_pickle=True)
```

```python
# 读取图片
image_path = r"D:\BaiduSyncdisk\jupyter\012_人脸识别\dataset\test\冯提莫.jpg"
image  = cv2.imdecode(np.fromfile(image_path, dtype=np.uint8), -1)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 人脸检测
faces = detector.detect_faces(image)
faces[0]['box']

# 人脸识别
fig, ax = plt.subplots()
ax.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
for face in faces:
    x, y, w, h = face['box']
    face_roi = gray[y:y + h, x:x + w]
    label, confidence = face_recognizer.predict(face_roi)
    print(f'label = {all_labels[label]} with a confidence of {confidence}')
    # cv2.putText(image, str(all_labels[label]), (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.8, (0, 255, 0))
    plt.text(x, y, all_labels[label], color='#00ff00', fontsize=8, ha='left', va='bottom')
    # cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), thickness=2)
    rect = Rectangle((x, y), w, h, linewidth=1, edgecolor='#00ff00', facecolor='none')
    ax.add_patch(rect)
plt.show()
```

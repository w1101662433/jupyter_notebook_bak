```python
import torch
import os,sys,time
import cv2
import numpy as np
from torch.optim import Adam,SGD
from mtcnn import MTCNN
from matplotlib import pyplot as plt
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import torch.optim as optim
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定中文字体，如SimHei
plt.rcParams['axes.unicode_minus'] = False  # 解决负号'-'显示为方块的问题
```

```python
Base_path = r"D:\BaiduSyncdisk\jupyter\012_人脸识别\dataset\生成的人脸"
all_labels = sorted([label for label in os.listdir(Base_path)])
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
features = []
labels = []

for person in all_labels:
    path = os.path.join(Base_path, person)
    label = all_labels.index(person)
    # loop every image in the folder
    if person not in ['刘亦菲','冯提莫','杨颖',]:
        continue
    for img in os.listdir(path):
        img_path = os.path.join(path, img)
        # read the image from the path
         
        img_array  = cv2.imdecode(np.fromfile(img_path, dtype=np.uint8), -1)
        
        if img_array.shape[-1] != 3:
            continue
        
        img_array = cv2.resize(img_array, (224, 224))
        img_array = np.transpose(img_array, (2, 0, 1))
       
        
        features.append(img_array)
        labels.append(label)
```

```python
torch.cuda.is_available()
```

```python
class CustomDataset(Dataset):
    # 将X,y转入DataLoader
    # dataset = CustomDataset(features, labels)
    # batch_data = DataLoader(dataset=dataset, batch_size=64, shuffle=True, drop_last=False)
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.int64)

    def __getitem__(self, index):
        # 运行数据加载器时，会返回下面的值
        X = self.X[index]
        y = self.y[index]
        return X, y

    def __len__(self):
        # 返回数据集的大小
        return len(self.y)

batch_data = CustomDataset(features, labels)
batch_data = DataLoader(dataset=batch_data, batch_size=64, shuffle=True, drop_last=False)
```

```python
import torchvision.models as models
import torch.nn as nn

# 加载预训练模型
model = models.resnet50(pretrained=True).to(device)

# 修改最后的全连接层以匹配你的类别数
num_ftrs = model.fc.in_features
num_classes = 19
model.fc = nn.Linear(num_ftrs, num_classes).to(device)  # num_classes是你的类别数
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 训练模型
num_epochs = 9
for epoch in range(num_epochs):  # num_epochs 是迭代轮数
    print('epoch',epoch)
    for inputs, labels in batch_data:
        inputs = inputs.to(device)
        labels = labels.to(device)
        print(inputs.shape)
        optimizer.zero_grad() # 清零梯度
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()  # 反向传播
        optimizer.step()  # 更新参数
        print(loss)
```

```python
# 读取图片
image_path = r"D:\BaiduSyncdisk\jupyter\012_人脸识别\dataset\test\刘亦菲.jpg"
image  = cv2.imdecode(np.fromfile(image_path, dtype=np.uint8), -1)

# 人脸检测
detector = MTCNN()
faces = detector.detect_faces(image)
faces[0]['box']
```

```python
transform = transforms.Resize((224, 224))
from matplotlib.patches import Rectangle

# 人脸识别
fig, ax = plt.subplots()
ax.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
for face in faces:
    x, y, w, h = face['box']
    face_roi = image[y:y + h, x:x + w]
    # 应用 transform
    face_roi = torch.tensor(face_roi,dtype=torch.float32)
    face_roi = np.transpose(face_roi, (2, 0, 1))
    # print(face_roi.shape)
    face_roi = transform(face_roi)
    face_roi = face_roi.unsqueeze(0)
    # print(face_roi.shape)
    label = model(face_roi.to(device))
    
    label_id = torch.argmax(label)
    print('all_labels[label_id]',all_labels[label_id])
    # cv2.putText(image, str(all_labels[label]), (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.8, (0, 255, 0))
    plt.text(x, y, all_labels[label_id], color='#00ff00', fontsize=8, ha='left', va='bottom')
    # cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), thickness=2)
    rect = Rectangle((x, y), w, h, linewidth=1, edgecolor='#00ff00', facecolor='none')
    ax.add_patch(rect)
plt.show()
```
